```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../input/breast-cancer/breast cancer.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# dropping uneccesary columns
df.drop(['id','Unnamed: 32'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#now changing diagnosis values into int
df['diagnosis'] = [1 if each == 'M' else 0 for each in df.diagnosis]
```

```python
df.head(3)
```

```python
#creating two sets from the data of independent and dependent variables
x = df.drop('diagnosis', axis = 1)
y = df.diagnosis.values
```

```python
x
```

```python
y
```

```python
#performing standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(x)
```

```python
X = scaler.fit_transform(x)
```

Finding components required in PCA by plotting graph

The total number of points on x and y axes are 13; so we will take 13 components in the PCA

Applying PCA on the standardized values 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 13)
```

```python
pca.fit_transform(X)
```

```python
#creating new components 
new_comp = pca.fit_transform(X)
```

```python
#Creating a dataframe of new components 
df2 = pd.DataFrame(new_comp, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
```

```python
df2
```

Model building 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df2,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

Generating classification report to see the accuracy of prediction

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction, digits = 2))
```
